### A lot of graph for data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Đọc dữ liệu (giả sử dataframe đã có sẵn tên là data)
data['Date/Time'] = pd.to_datetime(data['Date/Time'])  # Chuyển đổi cột ngày giờ

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
plt.plot(data['Date/Time'], data['Temp_C'], color='blue', label="Temperature (°C)")

# Thêm tiêu đề và nhãn
plt.xlabel("Date/Time")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Over Time")
plt.legend()
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

##### X-Axis (Date/Time):
* The x-axis represents the date and time, ranging from early 2012 to early 2013.
* It shows a continuous timeline of recorded temperature data.
##### Y-Axis (Temperature in °C):
* The y-axis represents the temperature in degrees Celsius.
* The values range approximately from -30°C to over 30°C.
##### Trend Analysis:
* The temperature follows a cyclical pattern, likely representing seasonal variations.
* At the beginning of the year (January 2012), the temperature is quite low, often below 0°C.
* As time progresses towards mid-year (June-July 2012), the temperature increases significantly, reaching peaks above 30°C.
* After mid-year, the temperature gradually declines, indicating the approach of colder months.
* By the end of the year (December 2012 - January 2013), the temperature drops again, mirroring the pattern from early 2012.
##### Volatility & Variability:
* The data exhibits significant short-term fluctuations, suggesting rapid temperature changes within short intervals.
* The variability is more pronounced in colder months compared to the warmer period.

```python
sns.histplot(data['Temp_C'], bins=30, kde=True)
plt.title("Temperature Distribution")
plt.show()
```

##### X-Axis (Temp_C - Temperature in °C):
- The x-axis represents temperature values, ranging approximately from -25°C to 35°C.
- It captures the frequency of temperature occurrences within this range.
##### Y-Axis (Count - Frequency):
- The y-axis represents the count of occurrences for each temperature range.
- The highest frequency appears to be around 500–600 occurrences for some temperature values.
##### Distribution Shape:
- The histogram exhibits a bimodal distribution, meaning there are two peaks, suggesting two common temperature ranges.
- The first peak occurs near 0°C, indicating that many temperature readings were around freezing.
- The second peak occurs near 20°C, indicating a second common temperature range.
##### Kernel Density Estimation (KDE) Line:
- The smooth blue KDE curve overlays the histogram, showing the estimated probability density.
- The two peaks further confirm a bimodal nature, likely due to seasonal variations (winter and summer temperatures).
##### Spread & Variability:
- The temperature values are widely spread, spanning from approximately -25°C to 35°C.
- This wide range suggests the dataset includes data from both cold and warm periods, reinforcing the seasonal nature.

```python
sns.scatterplot(x=data['Temp_C'], y=data['Rel Hum_%'])
plt.xlabel("Temperature (°C)")
plt.ylabel("Relative Humidity (%)")
plt.title("Temperature vs Humidity")
plt.show()
```

##### Data Distribution and Trends
- The data points are widely scattered, showing a non-linear relationship between temperature and humidity.
- There appears to be an inverse correlation:
    At lower temperatures (below 0°C), relative humidity values are generally high (mostly above 60%).
    At higher temperatures (above 20°C), relative humidity shows greater variation, including values below 40%.
    Many points cluster near 100% humidity, which suggests frequent high-humidity conditions.
##### Possible Interpretation
- The negative correlation might indicate that colder temperatures are associated with higher humidity (e.g., winter conditions where cold air holds less moisture, leading to condensation).
- At higher temperatures, the air can hold more moisture, so relative humidity varies more and can drop significantly.
- This trend aligns with natural weather behavior, where warmer air tends to be drier in some regions.

```python
# Chuyển đổi cột Date/Time thành kiểu datetime
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Lấy giờ từ Date/Time
data['Hour'] = data['Date/Time'].dt.hour

# Tính nhiệt độ trung bình theo giờ
avg_temp_per_hour = data.groupby('Hour')['Temp_C'].mean()

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_temp_per_hour.index,hue=avg_temp_per_hour.index, y=avg_temp_per_hour.values, palette="coolwarm",legend=False)

# Thêm tiêu đề và nhãn
plt.xlabel("Hour of the Day")
plt.ylabel("Average Temperature (°C)")
plt.title("Average Temperature by Hour")
plt.show()
```

##### Temperature Trend Throughout the Day
- Lowest temperatures occur between midnight (0h) and early morning (6-7h).
- The bars during these hours are cooler (blue shades), indicating lower temperatures.
- Temperatures start rising after sunrise, around 8 AM to 12 PM.
- The bars become lighter in color, indicating an increase in temperature.
- The highest temperature is observed between 12 PM and 3 PM, where the bars are the lightest in the warm (reddish) palette.
- After 3 PM, temperatures gradually decline, with bars becoming darker in the evening.
- The cooling trend continues until midnight.
##### Interpretation
- The pattern follows a typical diurnal temperature cycle, where temperatures are lowest at night and peak in the afternoon.
- This trend is consistent with natural daily heating and cooling processes, driven by solar radiation.
- The color gradient (coolwarm palette) effectively highlights this transition.

```python
sns.scatterplot(x=data['Wind Speed_km/h'], y=data['Temp_C'], alpha=0.7)

# Thêm tiêu đề và nhãn
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Temperature (°C)")
plt.title("Scatter Plot of Wind Speed vs Temperature")
plt.show()
```

##### Observed Patterns and Trends
- Data Distribution:
    Most data points are densely clustered in the low to moderate wind speed range (0-40 km/h).
    There are fewer data points for higher wind speeds (>40 km/h).
- Relationship Between Wind Speed and Temperature:
    Higher temperatures (above 20°C) are mostly observed at low wind speeds.
    Lower temperatures (below 0°C) seem to occur at a wider range of wind speeds.
    As wind speed increases, temperatures appear more spread out, with more occurrences of colder temperatures.
- Outliers:
    A few points show high wind speeds (above 60 km/h) but do not follow a clear trend.
##### Interpretation
- The general trend suggests that higher temperatures tend to occur when wind speeds are low, while lower temperatures are more evenly distributed across different wind speeds.
- This could indicate a cooling effect of wind, where higher wind speeds are associated with colder temperatures.
- The spread of points suggests no strong linear correlation between wind speed and temperature, but a possible negative relationship, meaning stronger winds might be associated with cooler temperatures.
##### Possible Real-World Implications
- Meteorological Insight:
    Wind can contribute to cooling, which is why stronger winds are often observed in colder conditions.
    Calm conditions (low wind speed) tend to retain heat, leading to higher temperatures.
- Weather Forecasting:
    This scatter plot can help in predicting temperature drops based on wind speed increases.

```python
# Đếm số lần xuất hiện của từng điều kiện thời tiết
weather_counts = data['Weather Condition'].value_counts()

# Tính tổng số lần xuất hiện để tính phần trăm
total_count = weather_counts.sum()

# Xác định ngưỡng 4% (các điều kiện có tỷ lệ nhỏ hơn 4% sẽ được gộp vào "Others")
threshold = 0.04 * total_count

# Tạo danh sách mới, gộp những điều kiện nhỏ thành "Others"
weather_counts_grouped = weather_counts.copy()
weather_counts_grouped[weather_counts < threshold] = 0  # Đặt giá trị nhỏ hơn ngưỡng về 0
weather_counts_grouped["Others"] = weather_counts[weather_counts < threshold].sum()  # Cộng dồn vào "Others"
weather_counts_grouped = weather_counts_grouped[weather_counts_grouped > 0]  # Loại bỏ các giá trị 0

# Vẽ biểu đồ tròn
plt.figure(figsize=(8, 8))
plt.pie(weather_counts_grouped, labels=weather_counts_grouped.index, autopct='%1.1f%%', 
        startangle=90, colors=plt.cm.Paired.colors)

# Thêm tiêu đề
plt.title("Weather Condition Distribution")

# Hiển thị biểu đồ
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x=data['Temp_C'], y=data['Rel Hum_%'], scatter_kws={"alpha":0.5}, line_kws={"color":"red"})

# Thêm tiêu đề và nhãn
plt.xlabel("Temperature (°C)")
plt.ylabel("Relative Humidity (%)")
plt.title("Temperature vs Relative Humidity with Regression Line")
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

##### Observed Trends
- Data Distribution:
    The data points form a dense cloud, covering a broad range of temperatures and humidity values.
    Most points are concentrated in the higher humidity range (60%-100%).
- Negative Correlation:
    The red regression line has a downward slope, indicating a negative correlation between temperature and relative humidity.
    As temperature increases, relative humidity tends to decrease.
- Variability:
    While the trend is negative, there is still significant spread in the data, meaning temperature is not the only factor affecting humidity.
##### Interpretation
- Physical Explanation:
    Warmer air can hold more moisture, so the relative humidity drops if the actual moisture content remains constant.
    Conversely, at lower temperatures, air can hold less moisture, leading to higher relative humidity.
- Weather Patterns:
    Cold conditions (low temperatures) are often associated with high humidity, such as in foggy or snowy weather.
    Hot conditions (high temperatures) usually bring lower humidity, which is common in dry, arid climates.
